```python
import pandas as pd
import matplotlib.pyplot as plt

house_alias = 'oak'

for house_alias in ['beech', 'oak', 'fir', 'maple', 'elm']:

    df = pd.read_csv(f'pico_reliability_data_{house_alias}.csv')
    df['hour_start'] = pd.to_datetime(df['hour_start_ms'], unit='ms')

    times = list(df['hour_start'])
    relay1_activations = list(df['relay1_activations'])
    relay1_all_activations = list(df['relay1_all_activations'])

    plt.figure(figsize=(12,4))
    plt.plot(times, relay1_all_activations, label='All activations', alpha=0.5)
    plt.plot(times, relay1_activations, label='Activations not after reboot', alpha=1)
    plt.title(f'{house_alias.capitalize()} relay1 activations per hour\nmax {max(relay1_activations)}, avg: {round(sum(relay1_activations)/len(relay1_activations),1)}')
    plt.ylabel(f'Activations per hour')
    plt.legend()
    plt.ylim(-5,100)
    plt.show()
```
